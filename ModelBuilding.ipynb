```python
import pandas as pd
```

```python
resume_data = pd.read_csv('Cleaned_Resumes.csv')
resume_data
```

```python
X = resume_data['Resume_Details'].values
X
```

```python
Y = resume_data['Category'].values
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, random_state=45, test_size=0.25)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vector = TfidfVectorizer(sublinear_tf=True,stop_words='english')
x_train = tfidf_vector.fit_transform(x_train)
x_test = tfidf_vector.transform(x_test)
```

```python
x_train.shape, x_test.shape
```

```python
print(y_train)
```

```python
print(y_test)
```

# Classification Models

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, classification_report, precision_score, recall_score
from sklearn.model_selection import cross_val_score
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)

model_knn.fit(x_train, y_train)
y_pred = model_knn.predict(x_test)

accuracy_knn = accuracy_score(y_test, y_pred)

print(f'Accuracy of training set: {model_knn.score(x_train, y_train):}')

print(f'Accuracy of test set: {model_knn.score(x_test, y_test):}')

print(f"Classification report for classifier {model_knn}:\n{classification_report(y_test, y_pred)}\n")

precision_knn = round(precision_score(y_test, y_pred, average='macro'), 2)
recall_knn = round(recall_score(y_test, y_pred, average='macro'), 2)
f1_knn = round(f1_score(y_test, y_pred, average='macro'), 2)
accuracy_knn = round(accuracy_score(y_test, y_pred), 2)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold, cross_val_score
```

```python
model_DT = DecisionTreeClassifier(criterion='gini', max_depth = 3)

model_DT.fit(x_train, y_train)
y_pred = model_DT.predict(x_test)

accuracy_DT = accuracy_score(y_test, y_pred)

print(f'Accuracy of training set: {model_DT.score(x_train, y_train):.2f}')

print(f'Accuracy of test set: {model_DT.score(x_test, y_test):.2f}')

print(f"Classification report for classifier {model_DT}:\n{classification_report(y_test, y_pred)}\n")

precision_DT = round(precision_score(y_test, y_pred, average='macro'), 2)
recall_DT = round(recall_score(y_test, y_pred, average='macro'), 2)
f1_DT = round(f1_score(y_test, y_pred, average='macro'), 2)
accuracy_DT = round(accuracy_score(y_test, y_pred), 2)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF = RandomForestClassifier(n_estimators=200,max_samples = 0.25)

model_RF.fit(x_train, y_train)

y_pred = model_RF.predict(x_test)

accuracy_RF = accuracy_score(y_test, y_pred)

print(f'Accuracy of training set: {model_RF.score(x_train, y_train):.2f}')

print(f'Accuracy of test set: {model_RF.score(x_test, y_test):.2f}')

print(f"Classification report for classifier {model_RF}:\n{classification_report(y_test, y_pred)}\n")

precision_RF = round(precision_score(y_test, y_pred, average='macro'), 2)
recall_RF = round(recall_score(y_test, y_pred, average='macro'), 2)
f1_RF = round(f1_score(y_test, y_pred, average='macro'), 2)
accuracy_RF = round(accuracy_score(y_test, y_pred), 2)
```

# Support Vectore Machine (SVM) Classifier

# Train and Test Accuracy Plot

```python
import matplotlib.pyplot as plt
import numpy as np

model_names = ["KNN", "Decision T", "Random F", "SVM ", "Logistic", "Bagging", "Grad Boost","Naive Bayes"]
Test_accuracies = [accuracy_knn*100, 
                   accuracy_DT*100, 
                   accuracy_RF*100, 
                   accuracy_svm*100, 
                   accuracy_lgr*100, 
                   accuracy_bagg*100, 
                   accuracy_GradientBoost*100, 
                   accuracy_NB*100]
Train_accuracies = [model_knn.score(x_train, y_train)*100, 
                    model_DT.score(x_train, y_train)*100, 
                    model_RF.score(x_train, y_train)*100, 
                    model_svm.score(x_train, y_train)*100, 
                    model_lgr.score(x_train, y_train)*100, 
                    model_bagg.score(x_train, y_train)*100, 
                    model_GradientBoost.score(x_train, y_train)*100, 
                    model_NB.score(x_train, y_train)*100]

plt.figure(figsize=(12, 6), dpi=500)
x_pos = np.arange(len(model_names))
bar_width = 0.4

plt.bar(x_pos - bar_width/2, Train_accuracies, width=bar_width, label='Train', color="orange")
plt.bar(x_pos + bar_width/2, Test_accuracies, width=bar_width, label='Test', color="blue")

plt.xlabel("Classifiers", fontsize=16, fontweight='bold')
plt.ylabel("Accuracy", fontsize=16, fontweight='bold')
plt.title("Model Accuracy Scores", fontsize=18, fontweight='bold')

plt.xticks(x_pos, model_names, rotation=10)
plt.legend()

for i, acc_train in enumerate(Train_accuracies):
    plt.text(x_pos[i] - bar_width/2, acc_train, f"{acc_train:.2f}%", ha='center', va='bottom', fontsize=7)

for i, acc_test in enumerate(Test_accuracies):
    plt.text(x_pos[i] + bar_width/2, acc_test, f"{acc_test:.2f}%", ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.show()
```
